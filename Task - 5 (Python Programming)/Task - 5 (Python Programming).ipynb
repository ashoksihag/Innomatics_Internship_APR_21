```python
#question:3 Group(), Groups() & Groupdict()

pattern = r'([a-z A-Z 0-9])\1'

import re

S = input()

m = re.search(pattern, S)

if m:
    print(m.groups()[0])
else:
    print(-1)
```

```python
#question:4 Re.findall() & Re.finditer()

import re
v='aeiou'
c='qwrtypsdfghjklzxcvbnm'
p = r'(?<=[%s])([%s]{2,})[%s]' % (c,v,c)
s= input()
m = re.findall(pattern = p , string=s, flags=re.IGNORECASE)
if m == []:
    print(-1)
else:
    for i in m:
        print(i)
```

```python
#question:5 Re.start() & Re.end()

S = input()
k = input()
import re
pattern = re.compile(k)
m =pattern.search(S)

if not m:
    print('(-1, -1)')
else:
    while m:
        print('({0}, {1})'.format(m.start(),m.end()-1))
        m=pattern.search(S, m.start()+1)
```

```python
#question:6 Regex Substitution

import re
n = int(input())

pattern = r'(?<= )(&&|\|\|)(?= )'
replace = lambda x: 'and' if x.group() == '&&' else 'or'

for i in range(n):
    s = input()
    print(re.sub(pattern, replace, s))
```

```python
#question:11 HTML Parser - Part 1

from html.parser import HTMLParser

#define class
class MyHTMLParser(HTMLParser):
    
    #handle_starttag
    def handle_starttag(self, tag, attrs):
        print('Start :', tag)
        for attr in attrs:
            print('->',' > '.join(map(str, attr)))
            
    #handle_endtag
    def handle_endtag(self, tag):
        print('End   :', tag)
        
    #handle_statrendtag
    def handle_startendtag(self, tag, attrs):
        print('Empty :', tag)
        for attr in attrs:
            print('->',' > '.join(map(str,attr)))
            
#getting all the html lines
html = ''
for i in range(int(input())):
    html+=input()
    
#defining the parser and executing
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

```python
#question:12 HTML Parser - Part 2

from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    #handle_comment
    def handle_comment(self, data):
        if data != '\n':
            if '\n' not in data:
                print('>>> Single-line Comment')
                print(data)
            else:
                print('>>> Multi-line Comment')
                print(data)
    #handle_data
    def handle_data(self, data):
        if data != '\n':
            print('>>> Data')
            print(data)
    
    
html = ""       
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

```python
#question:14 Validating UID

import re
T = int(input())
for i in range(T):
    UID = input().strip()
    if UID.isalnum() and len(UID)==10:
        if bool(re.search(r'(.*[A-Z]){2,}',UID)) and bool(re.search(r'(.*[0-9]){3,}',UID)):
            if re.search(r'.*(.).*\1+.*',UID):
                print('Invalid')
            else:
                print('Valid')
        else:
            print('Invalid')
    else:
        print('Invalid')
```

```python
#question:15 Validating Credit Card Numbers

import re
n = int(input())
for i in range(n):
    N = input().strip()
    validation = re.search(r'^[456]\d{3}(-?)\d{4}\1\d{4}\1\d{4}$',N)
    if validation:
        combine = ''.join(validation.group(0).split('-'))
        final_validation = re.search(r'(\d)\1{3,}', combine)
        if final_validation:
            print('Invalid')
        else:
            print('Valid')
    else:
        print('Invalid')
```
